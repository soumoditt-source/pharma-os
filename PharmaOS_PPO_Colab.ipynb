# PharmaOS PPO baseline on Colab T4

This notebook trains the real PPO baseline defined in `train_ppo.py` against the real PharmaOS environment. Set the runtime to GPU if available.

```python
import os
import subprocess
import sys

REPO_URL = os.environ.get("PHARMAOS_REPO_URL", "https://github.com/<your-username>/pharma-os.git")
REPO_DIR = "/content/pharma-os"

if not os.path.exists(REPO_DIR):
    subprocess.run(["git", "clone", REPO_URL, REPO_DIR], check=True)

os.chdir(REPO_DIR)
subprocess.run([sys.executable, "-m", "pip", "install", "--upgrade", "pip"], check=True)
subprocess.run([sys.executable, "-m", "pip", "install", "-r", "requirements.txt"], check=True)
print("Working directory:", os.getcwd())
```

```python
import torch
print("CUDA available:", torch.cuda.is_available())
```

```python
subprocess.run([
    sys.executable,
    "train_ppo.py",
    "--task",
    "qed_optimizer",
    "--timesteps",
    "2500",
    "--device",
    "cuda",
    "--output-dir",
    "trained_agents_colab",
], check=True)
```

Training artifacts are written to `trained_agents_colab/`. Download them from Colab or copy them to your own storage after the run.
